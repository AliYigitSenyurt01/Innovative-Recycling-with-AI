```python
import os 
import re
import random
import numpy as np
import pandas as pd 
import seaborn as sns
from PIL import Image
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

tf.random.set_seed(1)
random.seed(1)
np.random.seed(1)
```

```python
base_path = os.curdir + "\\dataset-resized"

# Creating a dictionary of the 4 classes
categories={}
i = -1

for dirname, _, filenames in os.walk(base_path):
    for filename in filenames:
        categories[i] = dirname.split('\\')[-1]        
        break 
    
    i += 1

print("All Categories:")
for key, value in categories.items():
    print(key, "|",value)
```

```python
train_df = pd.read_csv("train.csv")
validate_df = pd.read_csv("validate.csv")
test_df = pd.read_csv("test.csv")
```

```python
df_visualization = train_df.copy()
# Change the catgegories from numbers to names
df_visualization['category'] = df_visualization['category'].apply(lambda x:categories[x] )

plt.figure(figsize=(6,5))
df_visualization['category'].value_counts().plot.bar(x = 'count', y = 'category', color='skyblue' )
plt.xlabel("Garbage Classes", labelpad=14)
plt.ylabel("Images Count", labelpad=14)
plt.title("Count of images per class", y=1.02);
```

```python
plt.figure(figsize=(8,8))
for i in range(9):
    random_row = random.randint(0, len(train_df)-1)
    sample = train_df.iloc[random_row]
    random_image = tf.keras.utils.load_img(base_path + "\\" + sample['filename'])
    plt.subplot(330 + 1 + i)
    plt.title(categories[sample['category']])
    plt.imshow(random_image)
    plt.axis("off")

plt.show()
```

```python
def load_images(df):
    X = []
    Y = []
    
    for index, row in df.iterrows():
        image_path = base_path + "\\" + row['filename']
        img = Image.open(image_path).convert("RGB")
        img = img.resize((100, 100))  # Resize to your desired input size
        img_array = np.array(img)  # Convert to NumPy array
        X.append(img_array/255.0)
        Y.append(row['category'])
        del img, img_array
    return np.array(X), np.array(Y)
```

```python
trainX, trainY = load_images(train_df)
valX, valY = load_images(validate_df)
testX, testY = load_images(test_df)
```

```python
rand = random.randint(0,len(trainX))
plt.imshow(trainX[rand])
plt.title(categories[trainY[rand]])
plt.axis("off")
plt.show()
```

```python
print("Shape of trainX =",trainX.shape)
print("Shape of trainY = ",trainY.shape)
print("Shape of valX   =",valX.shape)
print("Shape of valY   = ",valY.shape)
print("Shape of testX  =",testX.shape)
print("Shape of testY  = ",testY.shape)
```

# Classification Models

```python
accuracies = {}
```

## KNN Model

```python
trainX_flt = trainX.reshape(trainX.shape[0], -1)
valX_flt = valX.reshape(valX.shape[0], -1)
testX_flt = testX.reshape(testX.shape[0], -1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score, confusion_matrix


accuracy_values = []

for k in range(1,20,2):

    knn_classifier = KNeighborsClassifier(n_neighbors=k, metric="euclidean")
    knn_classifier.fit(trainX_flt, trainY)
    y_pred = knn_classifier.predict(valX_flt)
    
    accuracy = accuracy_score(valY, y_pred)
    accuracy_values.append(accuracy)
    print(f"Accuracy for {k}NN = {accuracy}")
```

```python
plt.plot(range(1,20,2), accuracy_values, marker='o')
plt.title('kNN Accuracy for Different k Values')
plt.xlabel('k Value')
plt.ylabel('Accuracy')
plt.show()
```

```python
KNN = KNeighborsClassifier(n_neighbors=1)
KNN.fit(trainX_flt, trainY)

predY_KNN = KNN.predict(testX_flt)
accuracies["KNN Model"] = accuracy_score(testY, predY_KNN)
print(f"Test Accuracy of KNN Model = {accuracies['KNN Model']}")
```

```python
conf_mat = confusion_matrix(testY, predY_KNN)

f,ax=plt.subplots(figsize=(6,6))
sns.heatmap(conf_mat,annot=True, cbar=False, cmap="Purples")
ax.set_xticklabels(categories.values())
ax.set_yticklabels(categories.values())
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

```python
print(classification_report(testY, predY_KNN))
```

```python
error = list(np.where(testY != predY_KNN)[0])
plt.figure(figsize=(10,10))
samples = random.sample(error, 9)

for i in range(9):
    sample = samples[i]
    true_label = categories[testY[sample]]
    predicted_label = categories[predY_KNN[sample]]
    image = testX[sample]
    
    plt.subplot(330 + 1 + i)
    plt.title(f'True: {true_label}\n Predicted: {predicted_label}')
    plt.imshow(image)
    plt.axis("off")

plt.show()
```

## MLP Model

```python
# One hot encoding the labels
from sklearn.preprocessing import OneHotEncoder
one_hot_encoder = OneHotEncoder(sparse=False)

trainY = one_hot_encoder.fit_transform(trainY.reshape(-1,1))
valY = one_hot_encoder.fit_transform(valY.reshape(-1,1))
testY = one_hot_encoder.fit_transform(testY.reshape(-1,1))

trainY
```

```python
MLP = tf.keras.models.load_model('MLP_model')
```

```python
score = MLP.evaluate(testX, testY)
accuracies["MLP Model"] = score[1]
```

```python
predY_ = MLP.predict(testX)
predY_MLP = np.argmax(predY_,axis=1)
trueY = np.argmax(testY,axis=1)

conf_mat=confusion_matrix(trueY, predY_MLP)

f,ax=plt.subplots(figsize=(6,6))
sns.heatmap(conf_mat,annot=True, cbar=False, cmap="Purples")
ax.set_xticklabels(categories.values())
ax.set_yticklabels(categories.values())
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

```python
print(classification_report(np.argmax(testY,axis=1), predY_MLP))
```

```python
error = list(np.where(trueY != predY_MLP)[0])
plt.figure(figsize=(10,10))
samples = random.sample(error, 9)

for i in range(9):
    sample = samples[i]
    true_label = categories[trueY[sample]]
    predicted_label = categories[predY_MLP[sample]]
    image = testX[sample]
    
    plt.subplot(330 + 1 + i)
    plt.title(f'True: {true_label}\n Predicted: {predicted_label}')
    plt.imshow(image)
    plt.axis("off")

plt.show()
```
